# numpy로 PCA하기

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
standardizing = StandardScaler()
```

```python
feature_std = standardizing.fit_transform(train_feature)
```

```python
cov_mat = np.dot(feature_std.T, feature_std)/(feature_std.shape[0]-1)
```

```python
eigenvalue, eigenvector = np.linalg.eig(cov_mat)
```

```python
eigenvalue.sort()
eigen_descending = eigenvalue[::-1]
```

```python
eigen_descending[0:23]
```

```python
eigenvalue, eigenvector = np.linalg.eig(cov_mat)
```

```python
idx = []
for i in np.arange(23) : 
    ind = np.where(eigenvalue == eigen_descending[i])[0][0]
    idx.append(ind)
```

```python
princomp = np.dot(feature_std, eigenvector)
```

```python
princomp_df = pd.DataFrame(princomp)
```

```python
princomp_95 = princomp_df.loc[: , idx]
```

```python
np.sum(princomp_95.apply(np.var, axis=0))/31
```

```python
princomp_95.head()
```

# sklearn으로 PCA하기

```python
from sklearn.decomposition import PCA
```

```python
explained_variance = []
for i in np.arange(feature_std.shape[1])+1 : 
    n_comp = i
    pca = PCA(n_components = n_comp)
    pca_feature = pca.fit(feature_std)
    cumsum = np.sum(pca.explained_variance_ratio_)
    explained_variance.append(cumsum)
```

```python
import matplotlib.pyplot as plt
plt.plot(explained_variance)
plt.plot((0,30), (0.95, 0.95))
plt.show()
```
